```python
data["Age"].fillna(data["Age"].median(),inplace=True)
data["Income"].fillna(data["Income"].median())
data["Spending_Score"].fillna(data["Spending_Score"].median())
data["City"].fillna(data["City"].mode()[0])
print(data)
```

```python
print(data["City"].mode())
print(data["Gender"].mode())
```

```python
data["Gender"].fillna(data["Gender"].mode(),inplace=True)
data["City"].fillna(data["City"].mode(),inplace=True)
print(data)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.sort_values(by="Income",ascending=True)
```

```python
Q1=data["Spending_Score"].quantile(0.25)
Q3=data["Spending_Score"].quantile(0.75)
IQR=Q3-Q1

print(Q1,Q3,IQR)
```

```python
data=(data['Income']>Q1-1.5*IQR)&(data['Income']<Q3+1.5*IQR)
print(data)
```

```python
data=pd.get_dummies(data,columns=["City"],drop_first=True,dtype=int)
print(data)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
data[['Age']]=st.fit_transform(data[['Age']])
data[['Income']]=st.fit_transform(data[['Income']])
data.head()
```
